```
all_words = []
with open("Alz_data_cleaned_summary.txt") as f:
    lines = f.readlines()  
    print(len(lines))
    sent = []
    for line in lines[:200]:
      all_words.append(line.split())
```

```
import numpy as np
```

```
type(all_words), len(all_words)
```

```
from gensim.models import Word2Vec
model = Word2Vec(all_words, min_count=1)
```

```
X = model[model.wv.vocab]
```

```
X.shape
```

```
from nltk.cluster import KMeansClusterer
import nltk
```

```
NUM_CLUSTERS=25
kclusterer = KMeansClusterer(NUM_CLUSTERS, distance=nltk.cluster.util.cosine_distance, repeats=10)
assigned_clusters = kclusterer.cluster(X, assign_clusters=True)
print (assigned_clusters)
```

```
dic = {}
for i in range(25):
  if int(i) not in dic:
    dic[int(i)]=[]
  your_word_vector = np.asarray(X[int(i)], dtype="float32")
  word = model.most_similar(positive=[your_word_vector], topn=50)
  dic[int(i)].append(word)
print(dic)
```
